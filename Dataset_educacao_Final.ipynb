**Plotagem do Mapa de Calor**

```python
# Selecionando as linhas que têm determinado recurso...
# ...Depois armazenando a quantidade de linhas (escolas) que têm o recurso numa variável.
# Essa variável possui como coluna a contagem de escolas e, como rótulo, a sigla do estado.

cont_escolas = dados_educacao['SG_UF'].value_counts()

tem_lab_info = dados_educacao[dados_educacao['IN_LABORATORIO_INFORMATICA'] == 1]
lab_info = tem_lab_info['SG_UF'].value_counts() / cont_escolas

tem_lousa_dig = dados_educacao[dados_educacao['IN_EQUIP_LOUSA_DIGITAL'] == 1]
lousa_dig = tem_lousa_dig['SG_UF'].value_counts() / cont_escolas

tem_multimidia = dados_educacao[dados_educacao['IN_EQUIP_MULTIMIDIA'] == 1]
multimidia = tem_multimidia['SG_UF'].value_counts() / cont_escolas

tem_desk_aluno = dados_educacao[dados_educacao['IN_DESKTOP_ALUNO'] == 1]
desk_aluno = tem_desk_aluno['SG_UF'].value_counts() / cont_escolas

tem_pcport_aluno = dados_educacao[dados_educacao['IN_COMP_PORTATIL_ALUNO'] == 1]
pcport_aluno = tem_pcport_aluno['SG_UF'].value_counts() / cont_escolas

tem_tablet_aluno = dados_educacao[dados_educacao['IN_TABLET_ALUNO'] == 1]
tablet_aluno = tem_tablet_aluno['SG_UF'].value_counts() / cont_escolas

tem_net_aprend = dados_educacao[dados_educacao['IN_INTERNET_APRENDIZAGEM'] == 1]
net_aprend = tem_net_aprend['SG_UF'].value_counts() / cont_escolas

#Equipamentos = [lab_info, lousa_dig, multimidia, desk_aluno, pcport_aluno, tablet_aluno, net_aprend]
#colunas = ['lab_info', 'lousa_dig', 'multimidia', 'desk_aluno', 'pcport_aluno', 'tablet_aluno', 'net_aprend']
```

```python
net_aprend
```

```python
 # Aqui é criado um dataframe para cada contagem de escolas por recurso tecnológico
 # As colunas de cada dataframe são renomeadas de acordo com o recurso.

df_multimidia = pd.DataFrame(multimidia, columns=['SG_UF'])
df_multimidia = df_multimidia.rename(columns={'SG_UF': 'multimidia'})

df_net_aprend = pd.DataFrame(net_aprend, columns=['SG_UF'])
df_net_aprend = df_net_aprend.rename(columns={'SG_UF': 'net_aprend'})

df_desk_aluno = pd.DataFrame(desk_aluno, columns=['SG_UF'])
df_desk_aluno = df_desk_aluno.rename(columns={'SG_UF': 'desk_aluno'})

df_lab_info = pd.DataFrame(lab_info, columns=['SG_UF'])
df_lab_info = df_lab_info.rename(columns={'SG_UF': 'lab_info'})

df_pcport_aluno = pd.DataFrame(pcport_aluno, columns=['SG_UF'])
df_pcport_aluno = df_pcport_aluno.rename(columns={'SG_UF': 'pcport_aluno'})

df_lousa_dig = pd.DataFrame(lousa_dig, columns=['SG_UF'])
df_lousa_dig = df_lousa_dig.rename(columns={'SG_UF': 'lousa_dig'})

df_tablet_aluno = pd.DataFrame(tablet_aluno, columns=['SG_UF'])
df_tablet_aluno = df_tablet_aluno.rename(columns={'SG_UF': 'tablet_aluno'})


# ...Em seguida, combinamos tudo em um único dataframe...

df_combined = pd.concat([df_multimidia, df_net_aprend, df_desk_aluno, df_lab_info, df_pcport_aluno, df_lousa_dig, df_tablet_aluno], axis=1) * 100 #Multiplica-se por 100 para exibir em porcentagem
print(df_combined)

#Cria a planilha com o dataframe criado
df_combined.to_excel('planilha_heatmap.xlsx', index=True)
```

```python
#Ordenação do dataframe
df_combined['media'] = df_combined.mean(axis=1)
df_ordenado = df_combined.sort_values('media')

#Remove-se a coluna "média", uma vez que não é mais necessária
df_ordenado = df_ordenado.drop('media', axis=1)

print(df_ordenado)

#Aqui os dados são transpostos, para que no eixo Y fiquem os recursos e, no eixo X, os estados.
df_ordenado = df_ordenado.T
```

```python
# Caso queiramos plotar a partir do arquivo Excel, basta criar um dataframe com a linha abaixo.
#df = pd.read_excel('planilha_heatmap.xlsx')

#Ajuste do tamanho da figura
plt.figure(figsize=(20, 6))

#Plotagem do gráfico
ax = sns.heatmap(df_ordenado, annot=True, fmt=".2f", cmap='YlOrRd')
plt.title('Porcentagem de escolas que possuem recursos tecnológicos por estado')

# Exibir o mapa de calor
plt.show()
```

**Plotagem dos Gráficos de Radar**

```python
#Seleção das escolas e das colunas para a plotagem dos gráficos de radar

escolas = []

escola1 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'ESCOLA DE APLICACAO DO RECIFE'])
escolas.append(escola1)

escola2 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'ESCOLA PREPARATORIA DE CADETES DO AR'])
escolas.append(escola2)

escola3 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'COLEGIO TIRADENTES IJUI'])
escolas.append(escola3)

escola4 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'COLEGIO TIRADENTES PASSO FUNDO'])
escolas.append(escola4)

escola5 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'COLEGIO MILITAR DE FORTALEZA'])
escolas.append(escola5)

escola6 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'COLEGIO TIRADENTES SANTA MARIA'])
escolas.append(escola6)

escola7 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'IRMA AGOSTINA ETEC'])
escolas.append(escola7)

escola8 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'PRESIDENTE VARGAS ETEC'])
escolas.append(escola8)

escola9 = (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'ESCOLA DE APLICACAO PROFESSORA IVONITA ALVES GUERRA'])
escolas.append(escola9)

escola10= (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'ESCOLA TECNICA ESTADUAL CICERO DIAS'])
escolas.append(escola10)

escola11= (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'COLEGIO MILITAR DE SALVADOR'])
escolas.append(escola11)

escola12= (dados_educacao.loc[dados_educacao['NO_ENTIDADE'] == 'EEEP ADRIANO NOBRE'])
escolas.append(escola12)

df_escolas = pd.DataFrame() #Cria um dataframe vazio para alocar as linhas selecionadas

for escola in escolas:
  df_escolas = df_escolas.append(escola)

df_radar = df_escolas[['IN_LABORATORIO_INFORMATICA', 'IN_EQUIP_LOUSA_DIGITAL', 'IN_EQUIP_MULTIMIDIA', 'IN_DESKTOP_ALUNO', 'IN_BANDA_LARGA', 'IN_COMP_PORTATIL_ALUNO', 'IN_TABLET_ALUNO', 'IN_INTERNET_APRENDIZAGEM']]

df_radar
```

```python
df_radar.columns
```

```python
#Reordenando as colunas para a plotagem
df_radar = df_radar.reindex(columns=['IN_EQUIP_MULTIMIDIA', 'IN_LABORATORIO_INFORMATICA', 'IN_DESKTOP_ALUNO',
                                     'IN_INTERNET_APRENDIZAGEM', 'IN_BANDA_LARGA', 'IN_COMP_PORTATIL_ALUNO', 'IN_EQUIP_LOUSA_DIGITAL', 'IN_TABLET_ALUNO'])

rotulos = ['Equipamentos Multimídia - Projetor', 'Laboratório de Informática', 'Desktop para Alunos', 'Internet para Aprendizagem',
           'Internet Banda Larga', 'Comp. Portátil Alunos', 'Lousa Digital', 'Tablets para Alunos']

#Contando quantas escolas possuem o equipamento em questão

qt_equipamentos = df_radar.iloc[0:].sum()

qt_equipamentos
```

```python
#Criando o gráfico usando o Plotly
fig = go.Figure()
fig.add_trace(go.Scatterpolar(r = qt_equipamentos,
                              theta = rotulos,
                              fill = "toself",
                              name = "Contagem de Escolas",
                              showlegend = True,))

#Adicionando o título
fig.update_layout(
    title = "Recursos Tecnológicas das 12 Melhores Escolas no Ideb 2021 (Ensino Médio)",
    polar=dict(
        radialaxis=dict(visible=True)),
    width=1024,  # Largura da figura em pixels
    height=768,  # Altura da figura em pixels
    legend=dict(font=dict(size=14))
)
```
